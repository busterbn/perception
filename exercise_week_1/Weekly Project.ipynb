```python
import cv2
import numpy as np
import imutils
from matplotlib import pyplot as plt

bgr_img = cv2.imread("appletree.jpg")
b,g,r = cv2.split(bgr_img)
image = cv2.merge([r,g,b])
plt.imshow(image)
```

```python
# Convert image to HSV color space for easier color segmentation
hsv = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

# Plot the HSV channels side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(h, cmap='gray')
axes[0].set_title('Hue')
axes[1].imshow(s, cmap='gray')
axes[1].set_title('Saturation')
axes[2].imshow(v, cmap='gray')
axes[2].set_title('Value')
plt.show()
```

```python
gray = h.copy()
edged = cv2.Canny(gray, 30, 150)
plt.imshow(edged, cmap='gray')
```

```python
threshold = 150

threshold_value = 255

thresh = cv2.threshold(gray, threshold, threshold_value, cv2.THRESH_BINARY_INV)[1]
plt.imshow(thresh, cmap='gray')
```
